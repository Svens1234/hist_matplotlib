```python
nut_butter_prices={"Almond butter": 10,
                  "Peanut butter": 8,
                  "Cashew butter": 12,}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title='Butter Store',
      ylabel="Price")
```
